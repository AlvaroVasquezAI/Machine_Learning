<h2 style='text-align:center' id='Taxi-v3-Environment'><strong>3. Taxi-v3 environment</strong></h2>

<h3 style='text-align:center'><a href='https://gymnasium.farama.org/environments/toy_text/taxi/'>Documentation</a></h3>

<p>The <code>Taxi-v3</code> environment is a simulation where a taxi must pick up passengers and drop them off at specific locations within a 5x5 grid. This environment has the following characteristics:</p>

<h3>Locations</h3>
<p>There are four predefined locations in the grid where passengers can be picked up and dropped off. These locations are marked by colors:</p>
<ul>
    <li><strong>R</strong>: (0, 0) - Red</li>
    <li><strong>G</strong>: (0, 4) - Green</li>
    <li><strong>Y</strong>: (4, 0) - Yellow</li>
    <li><strong>B</strong>: (4, 3) - Blue</li>
</ul>

<h3>Passenger locations</h3>
<p>The passenger can be located at one of the predefined locations or inside the taxi:</p>
<ul>
    <li><strong>0</strong>: Red (0, 0)</li>
    <li><strong>1</strong>: Green (0, 4)</li>
    <li><strong>2</strong>: Yellow (4, 0)</li>
    <li><strong>3</strong>: Blue (4, 3)</li>
    <li><strong>4</strong>: In the taxi</li>
</ul>

<h3>Destination locations</h3>
<p>The passenger's destination can be one of the four predefined locations:</p>
<ul>
    <li><strong>0</strong>: Red (0, 0)</li>
    <li><strong>1</strong>: Green (0, 4)</li>
    <li><strong>2</strong>: Yellow (4, 0)</li>
    <li><strong>3</strong>: Blue (4, 3)</li>
</ul>

<h3>Actions</h3>
<p>The taxi can perform six possible actions:</p>
<ul>
    <li><strong>Action 0</strong>: Move south</li>
    <li><strong>Action 1</strong>: Move north</li>
    <li><strong>Action 2</strong>: Move east</li>
    <li><strong>Action 3</strong>: Move west</li>
    <li><strong>Action 4</strong>: Pick up a passenger</li>
    <li><strong>Action 5</strong>: Drop off a passenger</li>
</ul>

<h3>Rewards</h3>
<p>Rewards in the <code>Taxi-v3</code> environment are designed to incentivize the agent to complete its task efficiently:</p>
<ul>
    <li><strong>Reward of -1</strong>: Penalty for each step taken, encouraging efficiency by reducing the number of steps.</li>
    <li><strong>Reward of -10</strong>: Penalty for attempting to pick up or drop off a passenger at an incorrect location, discouraging incorrect actions.</li>
    <li><strong>Reward of 20</strong>: Reward for dropping the passenger off at the correct location, indicating that the agent has successfully completed its task.</li>
</ul>

<h3>Episode end conditions</h3>
<p>The episode can end under the following conditions:</p>
<ul>
    <li><strong>Successful Task Completion</strong>: The taxi successfully picks up the passenger and drops them off at the correct location, receiving a reward of 20.</li>
    <li><strong>Time Limit</strong>: The episode reaches the maximum number of steps, typically 200 steps, without successfully completing the task.</li>
</ul>

<h3>State encoding</h3>
<p>The state space consists of 500 discrete states. Each state is a combination of the taxi's location, the passenger's location, and the destination location. The state is encoded as follows:</p>
<ul>
    <li><strong>Taxi Position</strong>: 25 possible positions (5x5 grid)</li>
    <li><strong>Passenger Location</strong>: 5 possible locations (4 predefined locations or inside the taxi)</li>
    <li><strong>Destination Location</strong>: 4 possible destinations</li>
</ul>
<code>State = ((taxi_row * 5 + taxi_col) * 5 + passenger_location) * 4 + destination</code>



```python
import gym # type: ignore
import numpy as np # type: ignore
import random
import pandas as pd # type: ignore
import matplotlib.pyplot as plt # type: ignore
from matplotlib.colors import ListedColormap # type: ignore
from IPython.display import clear_output # type: ignore
import time
```

```python
# Initialize the environment
env = gym.make('Taxi-v3')

print("Environment created")
print(env)
```

<h2 style='text-align:center' id='Optimal-Policy'><strong>4. Optimal policy</strong></h2>
<p>The optimal policy is the strategy that the agent should follow to maximize accumulated rewards and minimize penalties. In Q-learning, the policy is defined by the Q-table, which stores Q-values for each state-action pair. The Q-value represents the expected reward of taking an action in a given state, followed by the best possible policy.</p>


<h2 style='text-align:center' id='Training-the-Agent'><strong>5. Training the agent</strong></h2>
<p>The agent is trained using the Q-learning algorithm, a method of reinforcement learning. The training process consists of the following steps:</p>

<h3>Step 1: Initialization</h3>
<p>We initialize the learning parameters (learning rate <code>alpha</code>, discount factor <code>gamma</code>, and exploration rate <code>epsilon</code>) and the Q-table with zeros. The learning rate controls how much the Q-values are adjusted with each update, the discount factor determines the importance of future rewards, and the exploration rate controls the probability of taking a random action.</p>

<h3>Step 2: Interaction with the environment</h3>
<p>In each episode, the agent performs actions based on its current policy. Initially, this policy might be random (exploration) to discover new strategies. Over time, the agent will exploit the best-known actions based on the current values in the Q-table.</p>

<h3>Step 3: Receiving rewards</h3>
<p>After each action, the agent receives a reward from the environment and observes the new state. The reward informs the agent about the quality of the action taken. Positive rewards incentivize correct actions, while penalties discourage suboptimal actions.</p>

<h3>Step 4: Updating the Q-table</h3>
<p>The agent uses the Q-learning update equation to adjust the values in the Q-table:</p>
<p><code>Q(s, a) ← (1 - α) Q(s, a) + α [reward + γ max<sub>a'</sub> Q(s', a')]</code></p>
<p>This equation adjusts the Q-value for the current state-action pair based on the immediate reward and the maximum expected value of future actions. The learning rate <code>alpha</code> determines how much new experiences influence the Q-value updates, and the discount factor <code>gamma</code> represents the importance of future rewards.</p>

<h3>Step 5: Repetition</h3>
<p>The process repeats for many episodes, allowing the agent to improve its policy based on accumulated rewards and Q-table updates. Over time, the agent explores less and exploits more, using the knowledge gained to make optimal decisions.</p>


<h3 style='text-align:center'><strong>Parameters</strong></h3>

```python
# Initialize parameters
alpha = 0.1 # learning rate
gamma = 0.7 # discount factor
epsilon = 0.1 # exploration rate

# Initialize Q-table
Q = np.zeros((env.observation_space.n, env.action_space.n))
```

<h3 style='text-align:center'><strong>Training agent</strong></h3>

```python
episodes = 10000

for i in range(episodes):
    state = env.reset()
    state = state if isinstance(state, int) else state[0]
    penalties, reward = 0, 0
    done = False
    passenger_picked = False
    passenger_dropped = False

    while not done and not passenger_dropped:
        if random.uniform(0,1) < epsilon:
            action = env.action_space.sample() # Explore action space
        else:
            action = np.argmax(Q[state]) # Exploit learned values
        
        step_result = env.step(action)

        if len(step_result) == 5:
            next_state, reward, terminated, truncated, info = step_result
            done = terminated or truncated
        else:
            next_state, reward, done, info = step_result

        next_state = next_state if isinstance(next_state, int) else next_state[0]
        old_value = Q[state, action]
        next_max = np.max(Q[next_state])
        new_value = (1 - alpha) * old_value + alpha * (reward + gamma * next_max)
        Q[state, action] = new_value
        state = next_state

        if reward == -10:
            penalties += 1

        # Verify if the passenger was picked up
        if action == 4 and reward == -1:
            passenger_picked = True
        
        # Verify if the passenger was dropped off
        if passenger_picked and action == 5 and reward == 20:
            passenger_dropped = True
        
    if i % 100 == 0:
        clear_output(wait=True)
        print(f"Episode: {i}")

print("Training finished.\n")
pd.DataFrame(Q).to_csv("Q_table.csv") 
```

<h2 style='text-align:center' id='Evaluating-the-Agent'><strong>6. Evaluating the agent</strong></h2>
<p>Once the agent has been trained, we evaluate its performance to see how efficiently it can complete its task. During evaluation, the agent always exploits the best-learned actions (without exploration). The evaluation process includes the following steps:</p>

<h3>Step 1: Initialization</h3>
<p>We reset the environment and initialize counters for epochs, penalties, and rewards. This prepares the environment for an unbiased evaluation of the agent's performance.</p>

<h3>Step 2: Interaction and data collection</h3>
<p>The agent performs actions based on the Q-values learned during training. In each episode, we monitor the agent's actions, rewards, and penalties. The key steps include:</p>
<ul>
    <li><strong>Action selection</strong>: The agent selects the best action for the current state based on the Q-table.</li>
    <li><strong>Environment response</strong>: The environment responds to the action, returning the next state, reward, and whether the episode has ended.</li>
    <li><strong>Update counters</strong>: We update the counters for epochs, rewards, and penalties based on the environment's response.</li>
    <li><strong>Task completion check</strong>: We check if the agent has picked up the passenger and successfully dropped them off at the correct location.</li>
</ul>

<h3>Step 3: Results analysis</h3>
<p>After all episodes are completed, we analyze the collected data to evaluate the agent's performance. Key metrics include:</p>
<ul>
    <li><strong>Average steps per episode</strong>: This measures the efficiency of the agent in completing the task.</li>
    <li><strong>Average penalties per episode</strong>: This indicates how often the agent makes mistakes.</li>
    <li><strong>Average rewards per episode</strong>: This reflects the overall performance and success of the agent.</li>
    <li><strong>Success rate</strong>: This measures the percentage of episodes where the agent successfully completes the task.</li>
</ul>

```python
Q = pd.read_csv("Q_table.csv", index_col=0).values
max_iterations_per_episode = 50
n_episodes = 200

total_epochs, total_penalties, total_rewards, successful_episodes = 0, 0, 0, 0
steps_per_episode = []
penalties_per_episode = []
rewards_per_episode = []

for episode in range(n_episodes):
    state = env.reset()
    state = state if isinstance(state, int) else state[0] 
    epochs, penalties, reward = 0, 0, 0
    done = False
    episode_rewards = 0
    passenger_picked = False  
    passenger_dropped = False 

    for iteration in range(max_iterations_per_episode):
        action = np.argmax(Q[state]) 
        step_result = env.step(action)
        if len(step_result) == 5:
            next_state, reward, terminated, truncated, info = step_result
            done = terminated or truncated
        else:
            next_state, reward, terminated, info = step_result

        next_state = next_state if isinstance(next_state, int) else next_state[0]

        if reward == -10:
            penalties += 1

        epochs += 1
        episode_rewards += reward
        state = next_state

        if action == 4 and reward == -1:
            passenger_picked = True

        if passenger_picked and action == 5 and reward == 20:
            passenger_dropped = True
            done = True

        if done and passenger_dropped:
            successful_episodes += 1
            break

    total_penalties += penalties
    total_epochs += epochs
    total_rewards += episode_rewards

    steps_per_episode.append(epochs)
    penalties_per_episode.append(penalties)
    rewards_per_episode.append(episode_rewards)

# Calculate metrics
average_steps_per_episode = total_epochs / n_episodes
average_penalties_per_episode = total_penalties / n_episodes
average_rewards_per_episode = total_rewards / n_episodes
success_rate = successful_episodes / n_episodes * 100

print(f"Results after {n_episodes} episodes:")
print(f"Average timesteps per episode: {average_steps_per_episode}")
print(f"Average penalties per episode: {average_penalties_per_episode}")
print(f"Average rewards per episode: {average_rewards_per_episode}")
print(f"Success rate: {success_rate}%")

# Create a DataFrame with the metrics
metrics_df = pd.DataFrame({
    'Episode': range(1, n_episodes + 1),
    'Steps': steps_per_episode,
    'Penalties': penalties_per_episode,
    'Rewards': rewards_per_episode
})

# Plot metrics
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.plot(metrics_df['Episode'], metrics_df['Steps'])
plt.xlabel('Episode')
plt.ylabel('Steps')
plt.title('Steps per Episode')

plt.subplot(2, 2, 2)
plt.plot(metrics_df['Episode'], metrics_df['Penalties'])
plt.xlabel('Episode')
plt.ylabel('Penalties')
plt.title('Penalties per Episode')

plt.subplot(2, 2, 3)
plt.plot(metrics_df['Episode'], metrics_df['Rewards'])
plt.xlabel('Episode')
plt.ylabel('Rewards')
plt.title('Rewards per Episode')

plt.subplot(2, 2, 4)
plt.plot(metrics_df['Episode'], [successful_episodes / (i + 1) * 100 for i in range(n_episodes)])
plt.xlabel('Episode')
plt.ylabel('Success Rate (%)')
plt.title('Cumulative Success Rate')

plt.tight_layout()
plt.show()

# Show summary metrics as a table
summary_metrics_df = pd.DataFrame({
    'Metric': ['Average Steps per Episode', 'Average Penalties per Episode', 'Average Rewards per Episode', 'Success Rate'],
    'Value': [average_steps_per_episode, average_penalties_per_episode, average_rewards_per_episode, success_rate]
})

# Show summary metrics as a table
plt.figure(figsize=(10, 4))
plt.axis('off')
plt.table(cellText=summary_metrics_df.values, colLabels=summary_metrics_df.columns, cellLoc='center', loc='center', colColours=['#f2f2f2']*2)
plt.title('Summary Metrics')
plt.show()
```
